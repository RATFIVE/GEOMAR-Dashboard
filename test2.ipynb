```python
from astroquery.jplhorizons import Horizons
import pandas as pd
from astropy.time import Time
import pandas as pd
import pytz

# Definiere die IDs der 8 Planeten im JPL Horizons System
planets = {
    "Mercury": 199, "Venus": 299, "Earth": 399, "Mars": 499,
    "Jupiter": 599, "Saturn": 699, "Uranus": 799, "Neptune": 899
}

# Zeitspanne: Letzte 2 Jahre, stündliche Schritte
epochs = {'start': '2023-02-14', 'stop': '2023-02-20', 'step': '1h'}

# Daten speichern
all_data = []

for planet, planet_id in planets.items():
    obj = Horizons(id=planet_id, location='500@sun', epochs=epochs)
    data = obj.vectors()

    # Speichere relevante Spalten
    df = data.to_pandas()#[["datetime_jd", "x", "y", "z", "vx", "vy", "vz"]]
    df["planet"] = planet  # Planetenname hinzufügen
    all_data.append(df)

# Alle Planeten in einem DataFrame zusammenführen
df_all = pd.concat(all_data, ignore_index=True)

# Speichern als CSV
#df_all.to_csv("planet_positions_last_2_years.csv", index=False)

# Umwandlung von Julianischem Datum in UTC-Zeit
df_all["datetime_utc"] = df_all["datetime_jd"].apply(lambda jd: Time(jd, format='jd').to_datetime())

# Umwandlung von UTC nach Berlin-Zeit
berlin_tz = pytz.timezone("Europe/Berlin")
df_all["datetime_berlin"] = df_all["datetime_utc"].apply(lambda dt: berlin_tz.fromutc(dt))

# delete datetime_utc and datetime_jd columns
df_all.drop(columns=["datetime_utc", "datetime_jd"], inplace=True)
df_all.rename(columns={"datetime_berlin": "datetime"}, inplace=True)

display(df_all)
```
